```python
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
latent_dim = 100  # Dimensão do vetor de ruído
num_classes = 10  # Classes do MNIST (0 a 9)
epochs = 5000  # Número de épocas
batch_size = 128  # Tamanho do batch
image_shape = (28, 28, 1)

# 🔹 Construção do Gerador
def build_generator():
    noise_input = layers.Input(shape=(latent_dim,))
    label_input = layers.Input(shape=(1,), dtype="int32")
    
    label_embedding = layers.Embedding(num_classes, latent_dim)(label_input)
    label_embedding = layers.Flatten()(label_embedding)
    
    merged = layers.multiply([noise_input, label_embedding])
    
    x = layers.Dense(128 * 7 * 7, activation="relu")(merged)
    x = layers.Reshape((7, 7, 128))(x)
    x = layers.Conv2DTranspose(64, kernel_size=3, strides=2, padding="same", activation="relu")(x)
    x = layers.Conv2DTranspose(1, kernel_size=3, strides=2, padding="same", activation="sigmoid")(x)

    return models.Model([noise_input, label_input], x, name="Generator")

# 🔹 Construção do Discriminador
def build_discriminator():
    image_input = layers.Input(shape=image_shape)
    label_input = layers.Input(shape=(1,), dtype="int32")
    
    label_embedding = layers.Embedding(num_classes, np.prod(image_shape))(label_input)
    label_embedding = layers.Flatten()(label_embedding)
    label_embedding = layers.Reshape(image_shape)(label_embedding)
    
    merged = layers.Concatenate()([image_input, label_embedding])
    
    x = layers.Conv2D(64, kernel_size=3, strides=2, padding="same", activation="relu")(merged)
    x = layers.Conv2D(128, kernel_size=3, strides=2, padding="same", activation="relu")(x)
    x = layers.Flatten()(x)
    x = layers.Dense(1, activation="sigmoid")(x)

    return models.Model([image_input, label_input], x, name="Discriminator")

# 🔹 Construção do modelo cGAN
def build_gan(generator, discriminator):
    discriminator.compile(loss="binary_crossentropy", optimizer=optimizers.Adam(learning_rate=0.0002, beta_1=0.5))
    
    noise_input = layers.Input(shape=(latent_dim,))
    label_input = layers.Input(shape=(1,))
    
    generated_image = generator([noise_input, label_input])
    discriminator.trainable = False
    
    validity = discriminator([generated_image, label_input])
    
    gan = models.Model([noise_input, label_input], validity, name="cGAN")
    gan.compile(loss="binary_crossentropy", optimizer=optimizers.Adam(learning_rate=0.0001, beta_1=0.5))
    
    return gan

# 🔹 Inicializar os modelos
generator = build_generator()
discriminator = build_discriminator()
gan = build_gan(generator, discriminator)

# 🔹 Carregar o dataset MNIST
(x_train, y_train), _ = tf.keras.datasets.mnist.load_data()
x_train = (x_train.astype(np.float32) / 255.0).reshape(-1, 28, 28, 1)

# 🔹 Função de treinamento do cGAN
def train(epochs, batch_size=128):
    half_batch = batch_size // 2
    
    for epoch in range(epochs):
        # Treinar o discriminador
        idx = np.random.randint(0, x_train.shape[0], half_batch)
        real_images = x_train[idx]
        real_labels = y_train[idx]
        
        noise = np.random.normal(0, 1, (half_batch, latent_dim))
        fake_labels = np.random.randint(0, num_classes, half_batch)
        fake_images = generator.predict([noise, fake_labels])
        
        d_loss_real = discriminator.train_on_batch([real_images, real_labels], np.ones((half_batch, 1)))
        d_loss_fake = discriminator.train_on_batch([fake_images, fake_labels], np.zeros((half_batch, 1)))
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
        
        # Treinar o gerador via GAN
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        sampled_labels = np.random.randint(0, num_classes, batch_size)
        valid_y = np.ones((batch_size, 1))

        g_loss = gan.train_on_batch([noise, sampled_labels], valid_y)

        # Exibir progresso
        if epoch % 500 == 0:
            print(f"Epoch {epoch} | D Loss: {d_loss:.4f} | G Loss: {g_loss:.4f}")
            generate_images(epoch)

# 🔹 Função para gerar imagens durante o treinamento
def generate_images(epoch, examples=10):
    noise = np.random.normal(0, 1, (examples, latent_dim))
    sampled_labels = np.arange(examples)
    generated_images = generator.predict([noise, sampled_labels])
    
    fig, axes = plt.subplots(1, examples, figsize=(10, 2))
    for i in range(examples):
        axes[i].imshow(generated_images[i, :, :, 0], cmap="gray")
        axes[i].axis("off")
    plt.show()

# 🔹 Compilar explicitamente o discriminador ANTES do treinamento
discriminator.compile(loss="binary_crossentropy",
                      optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001, beta_1=0.5))


# 🔹 Iniciar treinamento
train(epochs, batch_size)

# 🔹 Salvar o modelo treinado
generator.save("cgan_generator.h5")
discriminator.save("cgan_discriminator.h5")
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt

# Carregar o gerador salvo
generator = load_model("cgan_generator.h5")

# Dimensão do vetor de ruído (deve ser igual à usada no treinamento)
latent_dim = 100
```

```python
def generate_image(label):
    noise = np.random.normal(0, 1, (1, latent_dim))  # Gerar ruído aleatório
    label = np.array([label])  # Converter rótulo para array
    generated_image = generator.predict([noise, label])  # Gerar imagem
    
    # Exibir imagem gerada
    plt.imshow(generated_image.squeeze(), cmap="gray")
    plt.axis("off")
    plt.show()

# Gerar imagem de um dígito específico (ex: 3)
generate_image(3)
```

```python
def generate_multiple_images():
    num_examples = 10
    noise = np.random.normal(0, 1, (num_examples, latent_dim))  # Gerar ruído para cada número
    labels = np.arange(num_examples)  # Criar rótulos de 0 a 9

    generated_images = generator.predict([noise, labels])

    # Exibir todas as imagens
    fig, axes = plt.subplots(1, num_examples, figsize=(10, 2))
    for i in range(num_examples):
        axes[i].imshow(generated_images[i, :, :, 0], cmap="gray")
        axes[i].axis("off")
        axes[i].set_title(f"{labels[i]}")
    
    plt.show()

# Gerar uma amostra de cada número de 0 a 9
generate_multiple_images()
```
